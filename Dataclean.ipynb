```python
from pathlib import Path
import pandas as pd
import numpy as np


def read_data_tables():
    drifting_longlines = pd.read_csv('gfwdata/drifting_longlines.csv')
    drifting_longlines['geartype'] = 'drifting longlines'

    fixed_gear = pd.read_csv('gfwdata/fixed_gear.csv')
    fixed_gear['geartype'] = 'fixed gear'

    pole_and_line = pd.read_csv('gfwdata/pole_and_line.csv')
    pole_and_line['geartype'] = 'pole and line'

    purse_seines = pd.read_csv('gfwdata/purse_seines.csv')
    purse_seines['geartype'] = 'purse seines'

    trawlers = pd.read_csv('gfwdata/trawlers.csv')
    trawlers['geartype'] = 'trawlers'

    trollers = pd.read_csv('gfwdata/trollers.csv')
    trollers['geartype'] = 'trollers'

    unknown = pd.read_csv('gfwdata/unknown.csv')
    unknown['geartype'] = 'unknown'

    return [
        drifting_longlines, fixed_gear, pole_and_line, purse_seines, trawlers,
        trollers, unknown
    ]


names = [
    'drifting_longlines', 'fixed_gear', 'pole_and_line', 'purse_seines',
    'trawlers', 'trollers', 'unknown'
]
big_table = pd.concat(read_data_tables())
big_table['date'] = pd.to_datetime(big_table['timestamp'], unit='s')
conditions = [
    big_table['date'].dt.year == 2012, big_table['date'].dt.year == 2013,
    big_table['date'].dt.year == 2014, big_table['date'].dt.year == 2015,
    big_table['date'].dt.year == 2016
]

en_DJF_choices = [-0.9, -0.4, -0.4, 0.5, 2.5]
en_JFM_choices = [-0.7, -0.4, -0.5, 0.5, 2.1]
en_FMA_choices = [-0.6, -0.3, -0.3, 0.5, 1.6]
en_MAM_choices = [-0.5, -0.3, 0.0, 0.7, 0.9]
en_AMJ_choices = [-0.3, -0.4, 0.2, 0.9, 0.4]
en_MJJ_choices = [0.0, -0.4, 0.2, 1.2, -0.1]
en_JJA_choices = [0.2, -0.4, 0.0, 1.5, -0.4]
en_JAS_choices = [0.4, -0.3, 0.1, 1.9, -0.5]
en_ASO_choices = [0.4, -0.3, 0.2, 2.2, -0.6]
en_SON_choices = [0.3, -0.2, 0.5, 2.4, -0.7]
en_OND_choices = [0.1, -0.2, 0.6, 2.6, -0.7]
en_NDJ_choices = [-0.2, -0.3, 0.7, 2.6, -0.6]

big_table['en_DJF'] = np.select(conditions, en_DJF_choices, default=None)
big_table['en_JFM'] = np.select(conditions, en_JFM_choices, default=None)
big_table['en_FMA'] = np.select(conditions, en_FMA_choices, default=None)
big_table['en_MAM'] = np.select(conditions, en_MAM_choices, default=None)
big_table['en_AMJ'] = np.select(conditions, en_AMJ_choices, default=None)
big_table['en_MJJ'] = np.select(conditions, en_MJJ_choices, default=None)
big_table['en_JJA'] = np.select(conditions, en_JJA_choices, default=None)
big_table['en_JAS'] = np.select(conditions, en_JAS_choices, default=None)
big_table['en_ASO'] = np.select(conditions, en_ASO_choices, default=None)
big_table['en_SON'] = np.select(conditions, en_SON_choices, default=None)
big_table['en_OND'] = np.select(conditions, en_OND_choices, default=None)
big_table['en_NDJ'] = np.select(conditions, en_NDJ_choices, default=None)

filepath = Path('final_data_table.csv')
big_table.to_csv(filepath)
print(big_table.head(10))
```
